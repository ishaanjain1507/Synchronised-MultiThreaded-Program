```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([870139, 908543.466666667, 921048.5,895940.8,910530.766666667,883515.685714286,877131.775,873049.777777778,876396.54])
b = np.linspace(10, 50, a.size)
a = a/1000
plt.plot(b, a, "o-")
plt.ylim(700, 950)
plt.grid(True)
plt.title("Average time spent by a passenger in the tour")
plt.xlabel("Number of passengers")
plt.ylabel("Time spent in milliseconds")
plt.savefig("Plot 1.png")
```
